# Sensor thickness effect

The thickness of the active material of an X-ray sensor has multiple effects on the detection of incoming X-rays, if the X-rays are incident with an angle $\alpha_i$: 



X-rays are detected along the whole length $d$, which leads to the following effects:

1. The sensitivity of the detector increases with $\alpha_i$ since the path $d$ of the X-ray in the sensitive part of the detector increases as $d = d_0 / \cos(\alpha_i)$
2. Peaks shift along the $x$ direction due to the parallax effect. This effect is $\alpha_i$-dependent.
3. Peaks broaden since a X-ray beam is detected along the whole length $x_s$

From X-ray beam geometry shown in the figure above, it can be shown that for the transmitted intensity $I (\alpha_i)$ 

$$
I (\alpha_i) = \exp(- \mu d) \quad \text{with} \quad d = \frac{d_0}{\cos{\alpha_i}},
$$

and hence for the absorbtion efficiency $\Phi(\alpha_i)$ ($\approx$ quantum efficiency of the detector)

$$
\Phi (\alpha_i) = 1-  \exp(- \mu d)
$$


```python
%matplotlib widget
import matplotlib.pyplot as plt
from matplotlib import colors
import pyFAI
import xraydb
import numpy as np

az = pyFAI.load("./calibration_20keV.poni")

```

```python
thickness = 1000e-6 # Pilatus Si sensor thickness 
wavelength = az.get_wavelength()
dist = az.get_dist()
poni1 = az.get_poni1()
poni2 = az.get_poni2()
pixsize = az.get_pixel1()
energy = 45. #pyFAI.units.hc/(wavelength*1e10)
mu = xraydb.material_mu('Si',energy*1e3 ,kind='photo')*100.
eff = 1.0-np.exp(-mu*thickness)

print("µ = %f m^-1 hence absorption efficiency for 1000 µm: %.1f %% at normal incidence"%(mu, eff*100))
```

## 1. Sensitivity correction ($C_{d,I}$) 

Can be simply estimated from the ratio of absorbed X-rays $\Phi$ at $\alpha_i = 0$ and $\alpha_i \neq 0$:

$$
C_{d,I} = \frac{\Phi (\alpha_i) }{\Phi (0) } = \frac{1 - \exp(- \mu d)}{1 - \exp(- \mu d_0)} \quad \text{with} \quad d = \frac{d_0}{\cos{\alpha_i}} 
$$


## 2. Parallax correction

Parallax correction is complicated since the intensity along the detector plane $x$ is convoluted with an exponential function, which leads to broadening without a specific peak. The maximum shift is equal to $x_s = d \cdot \sin(\alpha_i)$, which corresponds to the width of the exponential used for the convolution. However, we can calculate the median shift $\Delta x$ of the intensity along $x$.

The total absorbed intensity through the sensor along the length $d$ is $\Phi(d)$ (neglect the $\alpha_i$-dependence for now). An approximation for the peak shift is the distance $d_{1/2}$, where half of the total dose is deposited. 

$$
\frac{1}{2} \Phi(d) = \Phi(d_{1/2}) \Rightarrow d_{1/2} = - \frac{\log ( \frac{1}{2} (1 + \exp(-\mu d)) )}{\mu}
$$




```python
fig = plt.figure()
ax = fig.add_subplot(111)

d = np.linspace(0,thickness,100)
I = np.exp(-mu*d)
ax.plot(d*1e6, I, 'k-')

Phi_d = I[-1]

I_median = (1 + Phi_d) / 2

d_median = - np.log(I_median) / mu

ax.plot(d_median*1e6,I_median, marker='o', color='r')
ax.set_ylabel(r'$I(y)$')
ax.set_xlabel(r'$y$ / $\mathrm{\mu}$m')

```

The displacement along the $x$ axis is then given by

$$
\Delta x = d_{1/2} \cdot \sin(\alpha_i)
$$

```python
fig = plt.figure()
ax = fig.add_subplot(111)

alpha_i = np.linspace(0, 60, 100)

for t, c in zip([450e-6, 1000e-6], ['k', 'b']):
    d = t / np.cos(np.deg2rad(alpha_i))
    
    
    
    #d = np.linspace(0,1400e-6,100)
    Phi_d = np.exp(-mu*d)

    I_median = (1 + Phi_d) / 2
    
    d_median = - np.log(I_median) / mu
    x_displ = np.sin(np.deg2rad(alpha_i)) * d_median
    
    ax.plot(alpha_i, (x_displ/pixsize) , ls='-', color=c)

ax.set_ylabel(r'$\Delta x$ / $\mathrm{\mu}$m')
ax.set_xlabel(r'$\alpha_i$ / $^{\circ}$')

#def I_mean(mu, dmax):
#    return (1 / (dmax*mu)) * (1 - np.exp(-mu*dmax))

#I_m = I_mean(mu, np.amax(d))

#d_m = - np.log(I_m) / mu



#ax.plot(d_median*1e6,I_median, marker='x', color='r')

```
